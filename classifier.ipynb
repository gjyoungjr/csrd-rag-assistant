```python
import fitz 

doc = fitz.open('./data/apple.pdf')


def chunk_text(text, max_chars=3000):
    paragraphs = text.split("\n\n")
    chunks, current_chunk = [], ""
    for paragraph in paragraphs:
        if len(current_chunk) + len(paragraph) <= max_chars:
            current_chunk += paragraph + "\n\n"
        else:
            chunks.append(current_chunk.strip())
            current_chunk = paragraph
    if current_chunk:
        chunks.append(current_chunk.strip())
    return chunks


for page in doc: 
    text = page.get_text()
    chunks = chunk_text(text)
    print (chunks)
```

```python
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model='gpt-4o-mini',
    temperature=0,
)


def classify_section(chunk):
    prompt = f"""
        Classify the following text from a sustainability report into its respective section (e.g., "Executive Summary", "Environmental Impact", "Social Responsibility", etc.):

        Text:
        {chunk}

        Response format:
        {{
        "Section": "Section Name",
        "Summary": "A brief summary of this section."
        }}
    """

    # Step 2: Define your messages
    messages = [
        {"role": "system", "content": "You are a helpful assistant."},
        {"role": "user", "content": prompt}
    ]

    # Step 3: Get a response from the model
    response = llm.invoke(messages)

    # Step 4: Print the response content
    print(response.content)
```

```python
for page in doc: 
    text = page.get_text()
    chunks = chunk_text(text)
    classify_section(chunks)
```
